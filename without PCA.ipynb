```python
from sklearn.preprocessing import LabelEncoder
object_cols = ['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities',
               'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
               'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
               'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
               'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
               'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
               'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual',
               'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature',
               'SaleType', 'SaleCondition']

le = LabelEncoder()
for col in object_cols:
    df[col] = le.fit_transform(df[col].astype(str))
```

```python
df.info()
```

```python
input=df.drop('SalePrice',axis=1)
output=df['SalePrice']
input
```

```python
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
x=st.fit_transform(input)
x
```

```python
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train=train_test_split(input,output,random_state=42,test_size=0.2)
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)                      
y_pred=model.predict(x_test)
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```
